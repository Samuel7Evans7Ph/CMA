```python
#question1 

import numpy as np
import random
np.set_printoptions(precision=8)
class Matrix():
    def __init__(self,a,b):
        number=0.0
        self.row=a
        self.col=b
        self.matrix=np.empty((a,b),dtype=np.float32)
        for i in range(a):
            for j in range(b):
                for_number="{:0.8f}".format(number)
                self.matrix[i,j]=for_number
                # print(for_number)
                # print(number)
                # print(self.matrix[i,j])
        

    # def __str__(self):
    #     col_len=len(self.matrix[0])
    #     row_len=len(self.matrix)
    #     # np.set_printoptions(precision=8)
    #     # print(self.matrix)
    #     print(self.matrix)
    #     # for i in range(row_len):
    #     #     for j in range(col_len):
    #     #         print(self.matrix[i,j],end=" ")
    #     #     print('\n')
    #     return ""
    def __str__(self):
        result = f'A {self.row} x {self.col} matrix with entries: \n'
        # print("hai")
        # print(self.matrix)
        for row in self.matrix:
            for val in row:
                result += f'{val : .8f}    '
            result += '\n'

        return result

    def __getitem__(self,index_list):
        return self.matrix[index_list]

    def __setitem__(self,index_list,value):
        self.matrix[index_list]=value

    def toEye(self):
        for row in range(self.row):
            for val in range(self.col):
                self.matrix[row,val]=0.000000
        for i in range(min(self.row,self.col)):
            self.matrix[i,i]=1.000000

    def toOne(self):
        for row in range(self.row):
            for val in range(self.col):
                self.matrix[row,val]=1.000000

    def randomise(self):
        for row in range(self.row):
            for val in range(self.col):
                self.matrix[row,val]=random.random()
        
        # return self.matrix
    def transpose(self):
        new_row=self.col
        new_col=self.row
        p2=Matrix(new_row,new_col)
        for i in range(new_row):
            for j in range(new_col):
                p2[i,j]=self.matrix[j,i]
        return p2

    def __add__(self,other):
        row_dim=other.row
        col_dim=other.col
        if(row_dim!=other.row or col_dim!=other.col):
            raise Exception ("not possible")
        else:
            new_matrix=Matrix(row_dim,col_dim)
            for  i in range(row_dim):
                for j in range(col_dim):
                    new_matrix[i,j]=self.matrix[i,j]+other.matrix[i,j]
            return new_matrix
    def __add__(self,other):
        row_dim=other.row
        col_dim=other.col
        if(row_dim!=self.row or col_dim!=self.col):
            raise Exception ("not possible")
        else:
            new_matrix=Matrix(row_dim,col_dim)
            for  i in range(row_dim):
                for j in range(col_dim):
                    new_matrix[i,j]=self.matrix[i,j]+other.matrix[i,j]
            return new_matrix

    def __sub__(self,other):
        row_dim=other.row
        col_dim=other.col
        if(row_dim!=self.row or col_dim!=self.col):
            raise Exception ("not possible")
        else:
            new_matrix=Matrix(row_dim,col_dim)
            for  i in range(row_dim):
                for j in range(col_dim):
                    new_matrix[i,j]=self.matrix[i,j]-other.matrix[i,j]
            return new_matrix


    def __mul__(self,other):
        print("hai")
        if not isinstance(other,int) and not isinstance(other,float):
            row_dim=other.row
            col_dim=other.col
            if(self.col!=other.row):
                raise Exception ("not possible")
            else:
                new_matrix=Matrix(self.row,col_dim)
                for  i in range(self.row):
                    for j in range(col_dim):
                        sum=0
                        for k in range(row_dim):
                            sum+=other.matrix[k,j]*self.matrix[i,k]
                        new_matrix[i,j]=sum
                return new_matrix
        else:
            new_matrix=Matrix(self.row,self.col)
            for i in range(new_matrix.row):
                for j in range(new_matrix.col):
                    new_matrix[i,j]=other*self.matrix[i,j]
            return new_matrix

        
      


    def __rmul__(self,other):
        print("nohai")
        if not isinstance(other,int) and not isinstance(other,float):
            row_dim=other.row
            col_dim=other.col
            if(col_dim!=self.row):
                raise Exception ("not possible")
            else:
                new_matrix=Matrix(self.row,col_dim)
                for  i in range(self.row):
                    for j in range(col_dim):
                        sum=0
                        for k in range(row_dim):
                            sum+=other.matrix[k,j]*self.matrix[i,k]
                        new_matrix[i,j]=sum
                return new_matrix
        else:
            new_matrix=Matrix(self.row,self.col)
            for i in range(new_matrix.row):
                for j in range(new_matrix.col):
                    new_matrix[i,j]=other*self.matrix[i,j]
            return new_matrix


    def norm(self,ord):
        transpose_matrix=self.transpose()
        col=np.array([np.linalg.norm(row,ord=ord) for row in transpose_matrix])
        return np.linalg.norm(col,ord=ord)


    def dominant_eigen(self):
        n=self.row
        p3=Matrix(n,1)
        # p3.randomise()
        for i in range(n):
            p3[i,0]=1
        for k in range(199):
            p3=self*p3
            print(p3)

            val=p3.norm(2)
            print(val,"val")
            a=1/val
            # a=a*a
            p3=p3*a
            print(p3)
        return p3
        
            
            
        
        


# m = Matrix(4, 4)
# m.randomize()
# m = m + m.t()
# eigvalues, eigvectors = np.linalg.eig(m.matrix)
# print(eigvalues)
# # while True:
# #     try:
# #         for val in eigvalues:
# #             if val < 0:
# #                 raise Exception('Algo will loop ???')
# #         m = m.randomize()
# #         m = m + m.t()
# #         break
# #     except:
# #         continue
# e, v = m.deflate()
        
        
        



    

    
    

    

    
        
        
        
        
        

                
m1=Matrix(3,4)
# m1[1,1]=3 def _str_(self):

# print(x)
# print(m1[1,1])
        
```
