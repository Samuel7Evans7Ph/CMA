```python
import matplotlib.pyplot as plt
import math as m
def fact(n):
    dp=[None]*(n+1)
    dp[0]=0
    for i in range(1,n+1):
        dp[i]=dp[i-1]+m.log(i)
    return dp
   
        
    
```

```python
#sterling approximation
def sterling(n):
    fac=[None]*(n+1)
    fac[0]=0
    for i in range(1,n+1):
        fac[i]=m.log(m.sqrt(2*m.pi*i))+i*m.log(i/m.e)
    return fac
   
```

```python
import numpy as np
n=100000
l1=fact(n)
l2=sterling(n)

c1=np.array([i for i in range(n+1)])
plt.plot(l1,c1)
plt.plot(l2,c1)
```

```python
#question 2
import random
class Dice:
    def __init__(self,sides=6):
        if type(sides)!=int or sides<4 or m.floor(sides)!=m.ceil(sides):
            raise Exception("Cannot construct the dice")
            
        self.sides=sides
    def simulate_roll(self,n):
        simulation=[0]*(self.sides)
        for i in range(n):
            t=random.random()
            for k in range(self.sides):
                if(t>k/(self.sides) and t<(k+1)/(self.sides)):
                    simulation[k]=simulation[k]+1
        actual_num=[i for i in range(1,self.sides+1)]
        for k in range(self.sides):
            simulation[k]/=n
        print(actual_num)
        print(simulation)
        fig = plt.figure(figsize =(10, 7))

        plt.bar(actual_num,simulation)
        plt.show()
        
        
    def set_prob(self,tup):
        prob_list=list(tup)
        self.prob=prob_list
#         print(prob_list)
        if(sum(prob_list)!=1) :
            raise Exception ("Invalid probability distribution")
        if(len(prob_list)!=self.sides):
            raise Exception("Probabilit for all faces not given")   
    def __str__(self):
            return f"dice with {self.sides} faces and probability distribution {self.prob}"
            
            
            
        
a=Dice(6)   
print(a.sides)
a.set_prob((0.1,0.2,0.3,0.1,0.1,0.2))
a.simulate_roll(10000)
#print(a.t)
print(a)
        
        
        
```
