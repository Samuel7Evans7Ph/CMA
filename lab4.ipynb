```python
#to visualise actual derivative
import matplotlib.pyplot as plt
import numpy as np
import math as m


#derivate of sin(x^2)
def calculate(h):
    itera=m.floor(1/h)
    list1=[i/itera for i in range(itera)]
    calc_ans1=[]
    actual_ans=[]
    for i in range(itera):
        actual_ans.append(np.cos(list1[i]**2)*2*list1[i])
        calc_ans1.append((np.sin((list1[i]+h)**2)-np.sin(list1[i]**2))/h)
    return actual_ans,calc_ans1    

actual_ans,calc_ans1=calculate(0.01)
plt.plot(list1,actual_ans)  
plt.plot(list1,calc_ans1)
    
```

```python

#question 2
def backward_diff1(h):
    itera=m.floor(1/h)
    calc_ans2=[]
    for i in range(itera):
        calc_ans2.append((np.sin(list1[i]**2)-np.sin((list1[i]-h)**2))/h)
    return calc_ans2     
    



def forward_diff(h):
    itera=m.floor(1/h)
    print(itera)
    abs_diff=[0]*itera
    actual_ans,calc_ans1=calculate(h)
    for i in range(itera):
        abs_diff[i]=np.abs(calc_ans1[i]-actual_ans[i])
    return abs_diff       

def backward_diff2(h):
    itera=m.floor(1/h)                     
    abs_diff2=[0]*itera
    actual_ans,calc_ans=calculate(h)
    calc_ans2=backward_diff1(h)
    for i in range(itera):
        abs_diff2[i]=np.abs(calc_ans2[i]-actual_ans[i])
    return abs_diff2  

def centrfin_diff(h):
    itera=m.floor(1/h)
    abs_diff3=[0]*itera
    abs_diff1=forward_diff(h)
    abs_diff2=backward_diff2(h)
    for i in range(itera):
        abs_diff3[i]=np.abs(abs_diff1[i]-abs_diff2[i])
    return abs_diff3

plt.plot(list1,forward_diff(0.01))
plt.plot(list1,backward_diff2(0.01))
plt.plot(list1,centrfin_diff(0.01))
```
