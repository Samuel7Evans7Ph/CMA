```python
import numpy as np
import matplotlib.pyplot as plt
from queue import Queue
class UndirectedGraph:
    
    def __init__(self,n=np.inf):
        self.nodes=n  
        self.edge=0
        self.adj_list={}
        if(type(n)==int):
            for i in range(n+1):
                self.adj_list[i]=[]
        
    def __add__(self,value):
        if isinstance(value,tuple):
            a,b=value
            self.adj_list[a].append(b)
            self.adj_list[b].append(a)
            self.edge+=1
        else:
            if value>self.nodes:
                raise Exception ("node index greater that the number of nodes")
            else:    
                print("hello")
                self.adj_list[value]=[]
                print(self.adj_list)
        return self    
            
            
            
        
        
            
    
    def add_node(self,n):
        if n>self.nodes:
            raise Exception ("node index greater that the number of nodes")
        else:
            self.adj_list[n]=[]
            
            
    def add_edge(self,a,b):
        self.adj_list[a].append(b)
        self.edge+=1
        self.adj_list[b].append(a)
        
    def call_from_str(self):
        print(f"graph with {self.nodes} nodes and {self.edge} edges.Neigbours of the nodes are belows")
        for i in self.adj_list:
             print(f"Node {i} :{self.adj_list[i]}")
        
        
    #to print an object of type undirected graph 
    def __str__(self):
        self.call_from_str()          
        return ""
        

#     to plot the degree distribution of the graph
    def plotDegDist(self):
        node_fract=[0]*(self.nodes+1)
        degree_list=[0]*(self.nodes+1)
        for i in range(self.nodes):
            
            node_fract[len(self.adj_list[i])]+=1
            degree_list[i]=i
        for i in range(self.nodes):
            node_fract[i]=node_fract[i]/self.nodes
        
        plt.scatter(degree_list,node_fract)
       
    
    #is connected
    def bfs(self,vis,start):
        count=0
        q=Queue()
        q.put(start)
        count=0
        while not q.empty():
            ele=q.get()
            count+=1
            vis[ele]=1
            for i in self.adj_list[ele]:
                if vis[i]==0:
                    q.put(i)      
        return count    
                
            
        
        
    
    
    def isconnected(self):
        #visited array
        vis=[0]*(self.nodes+1)
        self.bfs(vis,1)
        vis[0]=1
        if(vis!=[1]*(self.nodes+1)):
            print("NO")
            return False
        else:
            print("Yes")
            return True
    def oneTwoComponentSizes(self):
        second_max=0
        first_max=0
        vis=[0]*(self.nodes+1)
        for i in range(1,self.nodes+1):
            if vis[i]==0:
                siz=self.bfs(vis,i)
                if(siz>second_max):
                    if(siz>first_max):
                        first_max=siz
                    else:
                        second_max=siz
        return first_max,second_max                
                        
                
        
        
       
            
            
        
    
    
        
            
            
        
        
        
g=UndirectedGraph()   
g=UndirectedGraph(5)
print(g.nodes)
g=g+1
g=g+2
g=g+3
g=g+4
g=g+5
g=g+(2,3)
g=g+(4,5)
g=g+(4,1)
g=g+(4,2)
g=g+(3,4)
g=g+(1,5)
g.add_edge(1,2)
print(g)
g.plotDegDist()

        
        
```

```python
import random
import networkx as nx
import matplotlib.pyplot as plt
class Lattice:
    def __init__(self,n):
        self.n=n
        self.graph=self.create_lattice(n)
    def create_lattice(self,n):
        return nx.grid_2d_graph(n,n)
    def percolate(self,p):
        for edge in list(self.graph.edges()):
            if(random.random()>p):
                self.graph.remove_edge(*edge)
                
    def display(self):
        pos = {(x, y): (y, -x) for x, y in self.graph.nodes()}
        nx.draw(self.graph, pos, with_labels=True, node_color='skyblue', node_size=800, font_size=10)
        plt.title("Percolated Lattice Graph")
        plt.show()
                
            
    
        
        
    
    
```

```python
p1=Lattice(10)
```

```python
p1.percolate(0.65)
p1.display()
```
